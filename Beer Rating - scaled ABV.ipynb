# Training based on Review parameters

## Data Cleaning and Normalising

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

dataset = pd.read_csv('train.csv')

dataset.describe()
```

```python
dataset.head(5)
```

```python
#Scaling ABV value
dataset['standardized_abv'] = (dataset['beer/ABV']-dataset['beer/ABV'].mean())/dataset['beer/ABV'].std()
```

```python
dataset.head(5)
```

# Feature selection

```python
dataset.drop(['index','beer/ABV','standardized_abv','beer/brewerId','beer/name','beer/style','review/text','beer/beerId', 'user/ageInSeconds', 'user/birthdayRaw', 'user/birthdayUnix', 'user/gender', 'user/profileName', 'review/timeStruct', 'review/timeUnix'], axis=1, inplace=True)
```

```python
dataset.head(5)
```

```python
dataset.dropna( inplace=True)
```

```python
dataset.head(25)
```

# Training based on Review parameters

### StandardScaler - Feature Engineering

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split #cross

from sklearn.pipeline import make_pipeline
```

```python
y=dataset['review/overall'].values
X=dataset.drop('review/overall',axis=1).values
print(X)
```

# LINEAR REGRESSION

```python
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn import metrics
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
regressor = LinearRegression()  
regressor.fit(X_train, y_train) #training the algorithm
```

```python
y_pred = regressor.predict(X_test)
```

```python
df = pd.DataFrame({'Actual overall rating': y_test.flatten(), 'Predicted Overall Rating': y_pred.flatten()})
print(df.head(5))
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  

print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

## Lasso Regularization

```python
from sklearn.linear_model import Lasso

    
for alpha in [1.0, 0.1, 0.01, 0.001]:
    lasso = make_pipeline(StandardScaler(), Lasso(alpha=alpha, fit_intercept=True))
    #lassoreg = Lasso(alpha=alpha,normalize=True, max_iter=1e5)
    lasso.fit(X_train,y_train)
    y_pred = lasso.predict(X_test)
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  

print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

## Ridge Regularization

```python
from sklearn.linear_model import Ridge
# Ridge Regression Pipeline
for alpha in [1.0, 0.1, 0.01, 0.001]:
    ridge = make_pipeline(StandardScaler(), Ridge(alpha=alpha, fit_intercept=True))
    ridge.fit(X_train, y_train)
    y_pred = ridge.predict(X_test)
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  

print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

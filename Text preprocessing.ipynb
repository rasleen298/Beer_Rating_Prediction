# Text Preprocessing

```python
import string
import re
import nltk
from nltk.corpus import stopwords
import num2words

def clean_doc(review):
# split into tokens by white space
    #print(review)
    review=str(review)
    tokens = review.split()
    #print(tokens)
# prepare regex for char filtering
    re_punc = re.compile('[%s]' % re.escape(string.punctuation)) # remove punctuation from each word
    tokens = [re_punc.sub('', w) for w in tokens]
# remove remaining tokens that are not alphabetic
   # tokens = [word for word in tokens if word.isalpha()]
# filter out stop words
    stop_words = set(stopwords.words('english'))
    tokens = [w for w in tokens if not w in stop_words]
# filter out short tokens
    tokens = [word for word in tokens if len(word) > 1]
    
      
    #for i in tokens:
      #  if i.isdigit():
          #  i= num2words(i)
        
    return tokens
```

```python
def doc_to_line(tokens):
  
  tokens = clean_doc(tokens)
  # filter by vocab
  tokens = [w.lower() for w in tokens] 
  #lemmatizing and stemming
  #review = ' '.join([stemmer.stem(lemmatizer.lemmatize(word, pos='a')) for word in review.split()])
  return ' '.join(tokens)
  
```

```python
dataset['processed_text'] = dataset['review/text'].map(doc_to_line).astype(str)
dataset['processed_style'] = dataset['beer/style'].map(doc_to_line).astype(str)
dataset['processed_name'] = dataset['beer/name'].map(doc_to_line).astype(str)
```

```python
dataset.head(5)
```

```python
word_freq = dataset.processed_text.apply(lambda x : ' '.join(set(x.split()))).str.split(expand=True).stack().value_counts()
most_common_words = set(word_freq[ word_freq > 18000].keys()).difference({'not', 'sweet', 'flavor'})
least_common_words = set(word_freq[ word_freq < 4].keys())
# we shall remove both type of words and some others as necessary
undesirable_words = most_common_words.union(least_common_words).union({'drink', 'drinabl', 'and', 'like'})
```

```python
def remove_words(review):
    review=str(review)
    return ' '.join(list(filter(lambda word : word not in undesirable_words, review.split())))
```

```python
dataset['processed_text'] = dataset['processed_text'].map(remove_words)
```
